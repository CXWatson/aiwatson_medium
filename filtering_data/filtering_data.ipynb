# Imports

```python
import pandas as pd
```

# Reading in the CSV Data

```python
df = pd.read_csv("../iris.csv")
df
```

# Don't Do Section

```python
df.iloc[100:150]
```

# Do Section

```python
#Step 1
#Determine which column you want to get the values from
df["species"]
```

```python
#Step 2
#Select the column and perform a conditional statement
# which returns a series of true or false values
df["species"] == "virginica"
```

```python
#Step 3
# Use the above selection
# pass it back into the original DataFrame
# which will return the filtered selection
# Save the filter as a variable
virginica_filter = df["species"] == "virginica"
df[virginica_filter]
```

# Filtering with Numerical Data

```python
#Step 1
#Determine which column you want to get the values from
df["petal_width"]
```

```python
#Step 2
#Select the column and perform a conditional statement
# which returns a series of true or false values
df["petal_width"] > 1
```

```python
#Step 3
# Use the above selection
# pass it back into the original DataFrame
# which will return the filtered selection
# Save the filter as a variable
petal_width_greater_1_filter = df["petal_width"] > 1
df[petal_width_greater_1_filter]
```

# One-Liner Approach

```python
df[df["sepal_width"] > 3]
```

```python
# Extracting the filter
extract_filter = df["sepal_width"] > 3
df[extract_filter]
```
